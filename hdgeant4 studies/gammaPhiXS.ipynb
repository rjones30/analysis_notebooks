## gamma,phi diffractive photoproduction cross section as a function of gamma beam energy
These data were digitized visually from Fig. 3 in Wang et al, arXiv:2208.10289v1 [nucl-th] (22-Aug-2022)

```python
import ROOT
import numpy as np
```

```python
digiPoints = [
(0.044000, -100),
(0.044160, -0.249643),
(0.046590, 0.029382),
(0.048877, 0.233884),
(0.058802, 0.459170),
(0.079945, 0.590849),
(0.124724, 0.667028),
(0.242584, 0.720550),
(0.389838, 0.755758),
(0.527572, 0.776850),
(0.696672, 0.797753),
(0.861946, 0.814797),
(0.974966, 0.819801),
]

Egamma = []
gammaPhiXS_nb = []
for point in digiPoints:
    Egamma.append(point[0] * 21 + 1)
    gammaPhiXS_nb.append(10**(point[1] + 2))

Emin = 1.9
XSgraph = ROOT.TGraph(len(Egamma), np.array(Egamma, dtype=np.double), np.array(gammaPhiXS_nb, dtype=np.double))
XShisto = ROOT.TH1D('XShisto', "total phi(1020) photoproduction cross section", 220, 0, 22)
XShisto.GetXaxis().SetTitle("Egamma (GeV)")
XShisto.GetYaxis().SetTitle("#sigma_{tot} (GeV)")
for n in range(XShisto.GetNbinsX()):
    E = XShisto.GetXaxis().GetBinCenter(n+1)
    if E > Emin:
        XShisto.SetBinContent(n+1, XSgraph.Eval(E) * 1e-3)

c1 = ROOT.TCanvas("c1", "", 1200, 500)
c1.Divide(2,1)
pad1 = c1.cd(1)
XSgraph.Draw()
pad2 = c1.cd(2)
XShisto.SetStats(0)
XShisto.Draw()
c1.cd(0)
c1.Draw()
```
