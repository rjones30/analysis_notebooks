```python
import matplotlib
from datetime import datetime
import operator
import re

corehours = {}

nlines = 0
for line in open("sacct.listing"):
    field = line.rstrip().split()
    if len(field) == 9 and re.match(f"[0-9]", line):
        (jobid, start, end, state, partition, user, account, nodes, cpus) = field
        if state == "COMPLETED" and "2024" in start:
            try:
                tstart = datetime.strptime(start, "%Y-%m-%dT%H:%M:%S")
                tend = datetime.strptime(end, "%Y-%m-%dT%H:%M:%S")
                if not account in corehours:
                    corehours[account] = 0
            except:
                continue
            corehours[account] += (tend - tstart).total_seconds() * int(cpus) / 3600
            nlines += 1
print(f"nlines is {nlines}")

nlisted = 0
for act in sorted(corehours.items(), key=operator.itemgetter(1), reverse=True):
    nlisted += 1
    print(f"{nlisted}: {act[0]} - {corehours[act[0]]}")
```
