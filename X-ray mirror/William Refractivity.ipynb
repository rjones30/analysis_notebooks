```python
#function is n with parameter of wavelength in microns.
def lam(x):
    a = -1*3.5*10**-9
    b = 2.1835*10**-6
    c = -1*4.613988*10**-4
    d = 3.13828827*10**-2
    e = 2.4177539298
    f1 = -1*1.5606875731
    function = a*x**5 + b*x**4 + c*x**3 + d*x**2 + e*x+f1
    return function
#function is lambda in microns with parameter energy in eV
def g(E):
    """Takes in energy and returns lambda"""
    
    l = (1*10**-6)*(1.23984193*10**-6)/(E)
    
    return l
#function is n with parameter energy in eV
def n(E):
    return lam(g(E))
#function inputs wavelength in microns and outputs n1 of air
def air(x):
    n1 = 1 + 0.05792105/(238.0185-x**-2) + 0.00167917/(57.362-x**-2)
    return n1
#function inputs wavelength and outputs the critical angle in radians
def crit(x):
    n2 = lam(x)
    n1 = air(x)
    return np.arcsin(n2/n1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# Generate some data
x = np.linspace(1, 200, 1000)
y = lam(x)

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Set the x and y axis labels
ax.set_xlabel('E (eV)')
ax.set_ylabel('Refractivity Index, n')

# Set the plot title
ax.set_title('Energy (eV) vs Refractive Index')

# Show the plot
plt.show()
```
