```python
import numpy as np
from datetime import datetime
from matplotlib import pyplot as plt

overlog = "overhead-4-21-2023.log"

tseries = []
dseries = []
for line in open(overlog):
    sline = line.rstrip().split()
    time = datetime.strptime("{0}-{1:02d}-{2}T{3}".format(
                             sline[3], int(sline[4]), sline[7].rstrip(','),
                             sline[5]), "%b-%d-%YT%H:%M:%S")
    #if sline[3] != "Apr" or int(sline[4]) < 20 or (int(sline[5][:2]) < 14 and int(sline[4]) == 20):
    #    continue
    delay = int(sline[10]) / 60.
    tseries.append(time)
    dseries.append(delay)
 
# To draw scatter time series plot of the given dataframe
plt.plot_date(tseries, dseries)
 
# rotating the x-axis tick labels at 30degree towards right
plt.xticks(rotation=30, ha='right')
 
# Giving title to the chart using plt.title
plt.title('glidein startup delay (minutes) vs time')
 
# Providing x and y label to the chart
plt.xlabel('Date')
plt.ylabel('startup overhead (minutes)')
```
